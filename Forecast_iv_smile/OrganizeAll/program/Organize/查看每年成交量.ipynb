```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
start_year = 2012
end_year = 2023
years = np.arange(start_year, end_year+1, 1)
volumes = np.zeros(len(years))
top_path = './../../'
years_path = top_path + 'Data/Organized/year/'
year = 2012 
for i in range(len(years)):
    year = years[i]
    df = pd.read_csv('{}{}/TXO_{}.csv'.format(years_path, year, year), encoding='Big5', index_col=False)
    volume_year = df['成交量'].sum()
    volumes[i] = volume_year

plt.style.use('seaborn-darkgrid')

plt.bar(years, volumes)

plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
start_year = 2012
end_year = 2023
years = np.arange(start_year, end_year+1, 1)
volumes = np.zeros(len(years))
top_path = './../../'
years_path = top_path + 'Data/UnOrganized/'
year = 2014 
for year in years:
    year_path = '{}{}/opt/'.format(years_path, year)
    year_files = os.listdir(year_path)


    df = pd.read_csv('{}{}'.format(year_path, year_files[0]), encoding='Big5', index_col=False)
    year_data = pd.DataFrame(data=[], columns=df.columns.to_numpy())
    for year_file in year_files:
        df = pd.read_csv('{}{}'.format(year_path, year_file), encoding='Big5', index_col=False)
        year_data = pd.concat([year_data, df], axis=0)


    contracts = np.unique(year_data['契約'])
    volumes_contract = np.zeros(len(contracts))
    for i in range(len(contracts)):
        contract_data = year_data[year_data['契約'] == contracts[i]]
        volumes_contract[i] = contract_data['成交量'].sum()

    sort_idx = np.argsort(volumes_contract)
    #print(volumes_contract)
    x = contracts[sort_idx[-5:]]
    y = volumes_contract[sort_idx[-5:]]
    plt.style.use('seaborn-darkgrid')
    plt.bar(x, y)
    plt.title(year)
    plt.show()



#plt.style.use('seaborn-darkgrid')
#plt.bar(contracts, volumes_contract)
#plt.show()
```
