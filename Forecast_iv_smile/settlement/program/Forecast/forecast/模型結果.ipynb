```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt


top_path = './../../../'
model_path = top_path + 'Forecast&model/'
expiry = 'NearbyMonth'
#expiry = 'in30days'
IV_type = 'callIV'
K_Range = [300, 500]
K_Range_file = '{}_{}.csv'.format(K_Range[0], K_Range[1])
K_Range_dir = '{}_{}'.format(K_Range[0], K_Range[1])
seq_length1 = 7
min_contract = 2
model_type = 'LSTM-LSTM'
forecast_variable = 'slope'


model_Dir_tree = [expiry, IV_type, 'K_{}'.format(K_Range_dir), \
            'seq{}_min{}'.format(seq_length1, min_contract), model_type, forecast_variable]
for model_dir in model_Dir_tree:
    model_path = model_path + model_dir + '/'

ModellossData = pd.read_csv(model_path + 'Modelloss.csv', index_col=False, encoding='Big5')

ModellossMatrix = np.array(ModellossData.iloc[:-1, 1:])
ModellossMatrix[:, 0] = ModellossMatrix[:, 0].astype(int)
expiry = np.unique(ModellossMatrix[:, 0])
expiryMSE = np.zeros(len(expiry))

k = 8
for i in range(len(expiry)):
    Modelloss_expiry = ModellossMatrix[np.equal(ModellossMatrix[:,0], expiry[i])]
    ModelMSE_expiry = np.mean(Modelloss_expiry, axis=0)
    expiryMSE[i] = ModelMSE_expiry[k]

plt.style.use('seaborn-darkgrid')
plt.plot(expiry, expiryMSE, marker='o')
```
