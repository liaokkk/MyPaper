```python
import numpy as np
import pandas as pd
import os
s_c = '結算價'
top_path = './../../'
Data_path = top_path + 'Data/'
expiry = 'in30days'
#expiry = 'NearbyMonth'
IV_type = 'putIV.csv'
IV_path = '{}/{}/{}'.format(Data_path, expiry,  IV_type)
IV_data = pd.read_csv(IV_path, encoding='Big5', index_col=False)
IV_matrix = np.array(IV_data)


Dir_tree = [top_path, 'InterpData', expiry, IV_type[:-4]]
current_path = Dir_tree[0]
for i in range(1, len(Dir_tree), 1):
    if Dir_tree[i] not in os.listdir(current_path):
        os.mkdir(current_path + Dir_tree[i])
    current_path = current_path + Dir_tree[i] + '/'

dates = np.array(IV_data['交易日期'])
dates = np.unique(dates)
```

Udata將只包含與期貨訊息有關的資料，而不包含選擇權相關的資料

```python
column_names_Futures = ['交易日期', '到期日期', '期貨開盤價', '期貨最高價', '期貨最低價', '期貨收盤價',\
                        '期貨結算價', '到期天數', '無風險利率']
column_index_Futures = [IV_data.columns.get_loc(col) for col in column_names_Futures]
indices = [np.where(date == IV_matrix[:,0])[0][0] for date in dates]
Udata = IV_matrix[indices, :]
Udata = Udata[:, column_index_Futures]

```

將每天的資料按照履約價做篩選，篩出履約價位於結算價位指定範圍內的資料，並內插至100個點(流程為：如果邊界點(比方說指定前後500內，如果不存在，那就先靠內插得到點值)，然後篩出指定範圍的資料，篩出後，再將這個範圍內的點內插至100個點)。並挑一天畫出內插後的波動率曲線看看

```python
from function.interp import interpolate_range
import matplotlib.pyplot as plt
#買權
#upper_range = [300, 500, 800, 1000]
#lower_range = [0, 100, 200, 300, 500, 800]

#賣權
upper_range = [0, 100, 200, 300, 500, 800, 1000]
lower_range = [300, 500, 800, 1000]
K_num_day = 100

for i in range(len(upper_range)):
    for j in range(len(lower_range)):
        IV_K_newMatrix = np.zeros((len(dates), K_num_day , 2))
        file_name = '{}_{}.csv'.format(lower_range[j], upper_range[i])
        for k in range(len(dates)):
            one_day_matrix = IV_matrix[np.equal(IV_matrix[:, 0], dates[k])]
            F_day = one_day_matrix[0, IV_data.columns.get_loc('期貨{}'.format(s_c))]
            K_day = one_day_matrix[:, IV_data.columns.get_loc('履約價')]
            iv_day = one_day_matrix[:, IV_data.columns.get_loc('隱含波動率({})'.format(s_c))]

            K_upperbond = F_day + upper_range[i]
            K_lowerbond = F_day - lower_range[j]
            K_new_day, iv_new_day = interpolate_range(x=K_day, y=iv_day, \
                                          points=[K_lowerbond, K_upperbond], points_num=K_num_day)
            IV_K_newMatrix[k, :, :] = np.vstack((K_new_day, iv_new_day)).T
        InterpolateMatrix1 = np.reshape(np.array(['1','1']*(len(Udata)*K_num_day )), (-1,2))
        InterpolateMatrix2 = np.zeros((len(Udata)*K_num_day , len(column_names_Futures)-2+2))
        InterpolateMatrix = np.hstack((InterpolateMatrix1, InterpolateMatrix2))
        for k in range(len(Udata)):
            InterpolateMatrix[k*K_num_day :(k+1)*K_num_day , :9] = Udata[k]
            InterpolateMatrix[k*K_num_day :(k+1)*K_num_day , 9:] = IV_K_newMatrix[k]
        columns_all = np.array(column_names_Futures)
        columns_K_IV_interp = np.array(['履約價', '隱含波動率({})'.format(s_c),])
        columns_all = np.hstack((np.array(column_names_Futures), columns_K_IV_interp))
        InterpolateData = pd.DataFrame(data=InterpolateMatrix, columns=columns_all)
        InterpolateData.to_csv(current_path + file_name, encoding='Big5', index=False)

        l = 10
        x = IV_K_newMatrix[l, :, 0]
        y = IV_K_newMatrix[l, :, -1]
        K_min_l = y.min()
        K_max_l = y.max()
        plt.style.use('seaborn-darkgrid')
        plt.plot(x, y, c = [0.5, 0.5, 0.8], marker='o', label = 'IV Smile')
        plt.plot([Udata[l, 5]]*2, [K_min_l, K_max_l], ls='--', c=[0.8, 0.5, 0.5], label='Futures price')
        plt.legend()
        plt.show()
```
